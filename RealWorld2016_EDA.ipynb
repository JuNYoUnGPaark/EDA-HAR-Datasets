# RealWorld (HAR) 2016
---
- link: https://www.uni-mannheim.de/dws/research/projects/activity-recognition/dataset/dataset-realworld/  \
*(이유는 모르겠으나 VPN을 통해 download 가능)*
- 스마트폰, 스마트워치를 여러 몸 위치에 붙였을 때도 잘 동작하는 HAR + 위치인식 dataset
- 실내, 실외 혼합된 활동
- 15명 (8 male, 7 female)
- age = 31.9 ± 12.4
- height = 173.1 ± 6.9cm
- weight = 74.1 ± 13.8kg
---
- Activities (8가지)
1. `Climbingdown`
2. `Climbingup`
3. `Jumping`
4. `Lying`
5. `Running`
6. `Sitting`
7. `Standing`
8. `Walking`
---
- 센서 종류
1. **accelerometer (3-axis)**
2. **gyroscope (3-axis)**
3. magnetic field
4. light(조도)
5. GPS
6. sound level
---
- 센서 위치
1. chest (가슴)
2. forearm (전완)
3. head (머리)
4. shin (정강이)
5. thigh (대퇴골)
6. upper arm (팔뚝)
7. waist (허리)
---
- Sampling Rate = 50Hz(실제 계산결과 42Hz정도로 어느정도 오차있음)
- 각 활동을 피험자당 약 10분씩, jumping은 약 1.7분씩
---
- 특징
    1. 노이즈 및 변동성이 큰 편
    2. Multi-location 센서 구조 (단일도 가능)
    3. 전이구간 포함
    4. Cross-person variability

---
- `data/`: `.csv`, `.sqlite` 두 가지 종류
    - 각 data file내 `readMe`에 해당 데이터의 스펙 기재
    - 각 세부 `.csv`파일에는 id, attr_time(TS), x/y/z 기재
    - 상세 구조
        - `acc_climbingdown_csv.zip`
            - `acc_climbingdown_chest.csv`
            - `acc_climbingdown_forearm.csv`
            - ...
            - `readMe`
        - `acc_climbingup_csv.zip`
        - `acc_jumping_csv.zip`
        - ...
- `images/`: 센서 위치 다이어그램/설명 이미지
- `videos/`: 활동 촬영 영상

````python
import glob
import os

BASE_PATH = "/content/drive/MyDrive/Colab Notebooks/HAR_data/RealWorld2016"

# 우리가 찾을 대상 (7개 부위)
POSITIONS = ['chest', 'forearm', 'head', 'shin', 'thigh', 'upperarm', 'waist']
# 우리가 찾을 활동 (예시로 'walking' 하나만 먼저 찾아봅니다)
TARGET_ACTIVITY = 'walking'

def find_sensor_files(proband_id, activity):
    """
    특정 피실험자와 활동에 대한 Acc(7개), Gyro(7개) 파일 경로를 찾아서 반환합니다.
    """
    found_files = []

    # 찾을 센서 폴더 패턴 (acc와 gyr)
    sensor_types = ['acc', 'gyr']

    print(f"--- [Proband {proband_id}] Checking files... ---")

    for sensor in sensor_types:
        # 폴더 이름 예: acc_walking_csv 또는 gyr_walking_csv
        # Gyro는 폴더명이 'gyr_...' 이지만 파일명은 'Gyroscope_...' 로 시작함에 주의

        # 1. 먼저 해당 센서의 폴더 경로를 잡습니다.
        # proband1/data/acc_walking_csv 형태로 접근
        folder_pattern = os.path.join(BASE_PATH, f"proband{proband_id}", "data", f"{sensor}_{activity}_csv")

        # 2. 그 안에서 모든 csv 파일을 찾습니다 (recursive=True로 하위폴더까지 검색)
        # acc는 acc_*.csv, gyr는 Gyroscope_*.csv 패턴
        if sensor == 'acc':
            file_pattern = "acc_*.csv"
        else:
            file_pattern = "Gyroscope_*.csv"

        full_search_path = os.path.join(folder_pattern, "**", file_pattern)
        files = glob.glob(full_search_path, recursive=True)

        # 3. 찾은 파일 중에서 우리가 원하는 7개 부위만 골라냅니다.
        for pos in POSITIONS:
            # 파일 경로 안에 'chest', 'head' 등의 단어가 포함되어 있는지 확인
            # (대소문자 무시를 위해 lower() 사용)
            matched = [f for f in files if pos in f.lower()]

            if matched:
                # 여러 개가 잡히면(proband14 등) 첫 번째 것만 선택한다고 가정하고 출력
                found_files.append(matched[0])
                # print(f"  FOUND: {sensor.upper()} - {pos}") # 너무 길면 주석 처리
            else:
                print(f"  [MISSING] {sensor.upper()} - {pos} 파일을 찾을 수 없습니다.")

    return found_files

# =========================================================
# 실행: 1번부터 15번 사람까지 'walking' 파일 개수만 확인
# =========================================================
for i in range(1, 16): # 1~15
    files = find_sensor_files(i, TARGET_ACTIVITY)
    print(f"👉 Proband {i}: 총 {len(files)}개의 파일을 찾았습니다. (목표: 14개 Acc+Gyro)\n")
````

````python
import os
import glob
import pandas as pd

records = []
ROOT = "/content/drive/MyDrive/Colab Notebooks/HAR_data/RealWorld2016"

csv_paths = glob.glob(                                      # *: 무엇이든 들어와도 좋다
    os.path.join(ROOT, "proband*", "data", "**", "*.csv"),  # **: 중간에 폴더가 몇 개가 있든 상관없이 다 뚫고 지나가라
    recursive=True  # 재귀 모드 활성화
)

for full_path in csv_paths:
    # 1. 경로 분해
    rel = os.path.relpath(full_path, ROOT)  # 전체 경로에서 ROOT 부분을 떼어내기
    parts = rel.split(os.sep)  # / 기호를 기준으로 문자열을 잘라 리스트
    # ['proband1', 'data', 'acc_walking_csv', 'acc_walking_chest.csv']

    # 예외 처리: 파싱할 깊이가 안나오면 스킵 (readMe.txt)
    if len(parts) < 3:
        continue

    # 2. Proband ID 추출
    proband_str = parts[0]  # 'proband{i}'
    try:
        proband_id = int(proband_str.replace("proband", ""))
    except:
        continue # 폴더명이 이상한 경우 스킵

    # 3. Sensor & Activity 추출 (폴더명 기준)
    # 보통 parts[2]가 'acc_walking_csv' 형태
    # parts 구조가 [proband1, data, acc_..., file] 이므로 index 2가 맞음

    sensor_act_dir = parts[2]
    tokens = sensor_act_dir.split("_")

    # folder naming rule: {sensor}_{activity}_csv
    if len(tokens) >= 3:
        sensor = tokens[0]   # 'acc' or 'gyr' (폴더명 기준)
        activity = "_".join(tokens[1:-1]) # 'climbingup', 'walking'
    else:
        # 예외 케이스 처리
        sensor = "unknown"
        activity = sensor_act_dir

    # 4. Body Position & Trial 추출
    filename = parts[-1] # 파일명 (acc_walking_chest.csv)을 가져옵니다.
    stem = os.path.splitext(filename)[0] # 'acc_walking_chest' or 'acc_walking_2_chest'
    file_tokens = stem.split("_")

    # Body는 무조건 마지막 토큰 (chest, head...)
    body = file_tokens[-1]

    # Trial 추출 로직 강화
    # Case A: 폴더 깊이가 깊은 경우 (proband14/data/acc.../acc..._1_csv/file.csv)
    trial = 1 # 기본값

    if len(parts) == 5: # 구조가 proband / data / 활동폴더 / 세션폴더 / 파일 처럼 5단계인 경우
        trial_folder = parts[3] # 'acc_climbingup_1_csv'
        # 폴더명에서 숫자 추출
        t_nums = [int(s) for s in trial_folder.split("_") if s.isdigit()]
        if t_nums: trial = t_nums[0]

    # Case B: 파일명에 숫자가 있는 경우 (acc_walking_2_chest.csv)
    # 파일명 토큰 중 'acc', 'activity', 'body'를 제외하고 숫자가 남는지 확인
    # (간단하게 파일명에 숫자가 포함되어 있으면 그걸 trial로 봄)
    else:
        # 파일명 중간에 숫자가 끼어있는지 확인 (예: walking_2)
        for t in file_tokens:
            if t.isdigit():
                trial = int(t)
                break

    records.append({
        "proband_id": proband_id,
        "sensor": sensor,     # acc / gyr
        "activity": activity, # walking, running...
        "body": body,         # chest, head...
        "trial": trial,       # 1, 2, 3...
        "full_path": full_path
    })

# DataFrame 생성
meta = pd.DataFrame(records)

# 보기 좋게 정렬
meta = meta.sort_values(by=['proband_id', 'activity', 'trial', 'sensor', 'body'])

# 결과 확인
print(f"총 발견된 파일 수: {len(meta)}")
display(meta.head())
````

````python
# 그룹핑 기준: 사람(ID), 활동(Activity), 시도횟수(Trial)
grouped = meta.groupby(['proband_id', 'activity', 'trial'])

valid_sessions = []

print("데이터 세트 검증 중...")

for name, group in grouped:
    # name은 (proband_id, activity, trial) 튜플입니다.
    # group은 그 조건에 해당하는 파일들의 목록(DataFrame)입니다.

    # 조건: 센서 2종류(Acc/Gyr) * 7부위 = 총 14개 파일이어야 함
    if len(group) == 14:
        valid_sessions.append(name)
    else:
        # 14개가 아니면 뭔가 빠진 것 (로그 찍어보기)
        print(f"⚠️ [불완전 세트] ID:{name[0]}, Act:{name[1]}, Trial:{name[2]} -> 파일 {len(group)}개 뿐임 (Skip)")

print("-" * 30)
print(f"총 발견된 그룹 수: {len(grouped)}")
print(f"병합 가능한 완벽한 세트 수: {len(valid_sessions)}")
````

````python
from tqdm import tqdm

final_dataset = [] # 여기에 다 합친 df를 모을 예정

print("🚀 데이터 로딩 및 병합 시작 (Merge)...")

for (pid, act, trial) in tqdm(valid_sessions):  # 14개 파일이 모두 존재하는 세트만
    # 1. 현재 세션에 해당하는 파일 정보 가져오기
    current_files = meta[
        (meta['proband_id'] == pid) &
        (meta['activity'] == act) &
        (meta['trial'] == trial)
    ]

    # 2. 'acc_chest' 파일을 대장(Base)으로 선정
    # 가장 안정적이고 끊김이 적어서 보통 기준으로 사용
    # (조건에 맞는 행을 찾아서 full_path를 가져옴)
    base_row = current_files[(current_files['sensor'] == 'acc') & (current_files['body'] == 'chest')]

    if len(base_row) == 0:
        continue # 혹시 chest가 없으면 스킵

    base_path = base_row.iloc[0]['full_path']  # 대장 파일의 실제 경로-시리즈가 돼야 파일 읽기 가능

    # 대장 파일 읽기
    base_df = pd.read_csv(base_path)
    # 컬럼 정리 (id, time, x, y, z 가정)
    # 실제 컬럼명 확인 후 인덱스로 접근 (보통 [1]=time, [2,3,4]=x,y,z)
    base_df = base_df.rename(columns={
        base_df.columns[1]: 'time',
        base_df.columns[2]: 'acc_chest_x',
        base_df.columns[3]: 'acc_chest_y',
        base_df.columns[4]: 'acc_chest_z'
    })
    base_df = base_df[['time', 'acc_chest_x', 'acc_chest_y', 'acc_chest_z']]
    base_df = base_df.sort_values('time')

    # 3. 나머지 13개 파일 갖다 붙이기
    # (acc_chest를 제외한 나머지)
    other_rows = current_files[~((current_files['sensor'] == 'acc') & (current_files['body'] == 'chest'))]

    for _, row in other_rows.iterrows():
        # 파일 읽기
        temp_df = pd.read_csv(row['full_path'])

        # 컬럼명 만들기 (예: gyr_head_x)
        col_prefix = f"{row['sensor']}_{row['body']}" # gyr_head

        temp_df = temp_df.rename(columns={
            temp_df.columns[1]: 'time',
            temp_df.columns[2]: f'{col_prefix}_x',
            temp_df.columns[3]: f'{col_prefix}_y',
            temp_df.columns[4]: f'{col_prefix}_z'
        })
        temp_df = temp_df[['time', f'{col_prefix}_x', f'{col_prefix}_y', f'{col_prefix}_z']]
        temp_df = temp_df.sort_values('time')

        # 시간 기준 병합 (가장 가까운 시간)
        base_df = pd.merge_asof(base_df, temp_df, on='time', direction='nearest')

    # 4. 식별자 컬럼 추가
    base_df['subject_id'] = pid
    base_df['activity'] = act
    base_df['trial'] = trial

    final_dataset.append(base_df)

# 끝!
print("\n✅ 모든 처리 완료!")
````

````python
# 1. 리스트에 담긴 129개 데이터프레임을 하나로 합치기 (수직 결합)
master_df = pd.concat(final_dataset, ignore_index=True)

print("✅ 최종 병합 완료!")
print(f"👉 전체 데이터 크기 (Shape): {master_df.shape}")
print("-" * 30)

# 2. 데이터가 잘 섞여 들어갔는지 확인 (활동별 데이터 개수)
print("[활동별 데이터 샘플 수]")
print(master_df['activity'].value_counts())

# 3. 결측치(NaN) 확인
# (merge_asof 과정에서 시간 범위가 안 맞아 끝부분에 살짝 생길 수 있음)
null_counts = master_df.isnull().sum().sum()
print("-" * 30)
print(f"👉 총 결측치(NaN) 개수: {null_counts}")

if null_counts > 0:
    # 결측치가 있다면 앞뒤 값으로 채우거나 제거 (보통 아주 적으므로 제거해도 무방)
    master_df = master_df.dropna()
    print("   -> 결측치 제거 완료!")

# 4. 파일로 저장 (Pickle 포맷 추천 - 로딩 속도 1초 컷)
SAVE_PATH = "/content/drive/MyDrive/Colab Notebooks/HAR_data/RealWorld2016/RealWorld2016_Master_Sync.pkl"
master_df.to_pickle(SAVE_PATH)

print("-" * 30)
print(f"💾 파일 저장 완료: {SAVE_PATH}")
print("이제 분석 준비 끝입니다!")
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------------------------------------
# [설정] 1번 사람, Walking 데이터 전체 사용
# ---------------------------------------------------------
TARGET_ID = 5
TARGET_ACT = 'running'

# 데이터 필터링
subset = master_df[
    (master_df['subject_id'] == TARGET_ID) &
    (master_df['activity'] == TARGET_ACT)
].copy()

# ---------------------------------------------------------
# [분석] 가속도 X축끼리의 상관계수 계산
# ---------------------------------------------------------
# 분석할 컬럼들 (순서를 머리->발 순서로 배치해봅니다)
target_cols = [
    'acc_head_x', 'acc_chest_x', 'acc_waist_x',
    'acc_thigh_x', 'acc_shin_x',
    'acc_upperarm_x', 'acc_forearm_x'
]

# 상관계수 행렬 계산 (-1 ~ 1 사이 값)
corr_matrix = subset[target_cols].corr()

# 컬럼 이름 깔끔하게 정리 (acc_head_x -> Head)
clean_labels = [c.replace('acc_', '').replace('_x', '').upper() for c in target_cols]

# ---------------------------------------------------------
# [시각화] 히트맵 그리기
# ---------------------------------------------------------
plt.figure(figsize=(10, 8))

sns.heatmap(
    corr_matrix,
    annot=True,         # 숫자 표시
    fmt=".2f",          # 소수점 2자리
    cmap='coolwarm',    # 빨강(양의상관) ~ 파랑(음의상관)
    xticklabels=clean_labels,
    yticklabels=clean_labels,
    vmin=-1, vmax=1,    # 상관계수 범위 고정
    linewidths=1,       # 칸 구분선
    linecolor='white'
)

plt.title(f'Sensor Correlation Matrix (Subject {TARGET_ID} - {TARGET_ACT})', fontsize=16, pad=20)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# ---------------------------------------------------------
# [설정] 전체 데이터에서 '허벅지(Thigh)' 센서만 분석
# ---------------------------------------------------------
# 메모리 절약을 위해 필요한 컬럼만 복사
target_pos = 'thigh'
subset = master_df[['activity', f'acc_{target_pos}_x', f'acc_{target_pos}_y', f'acc_{target_pos}_z']].copy()

# ---------------------------------------------------------
# [계산] 가속도 크기 (Signal Vector Magnitude, SVM)
# ---------------------------------------------------------
# SVM = sqrt(x^2 + y^2 + z^2)
# 중력가속도(약 9.8)가 포함된 값이므로, 정지 상태여도 9.8 근처가 나옵니다.
subset['svm'] = np.sqrt(
    subset[f'acc_{target_pos}_x']**2 +
    subset[f'acc_{target_pos}_y']**2 +
    subset[f'acc_{target_pos}_z']**2
)

# ---------------------------------------------------------
# [시각화] 활동별 Boxplot
# ---------------------------------------------------------
plt.figure(figsize=(14, 8))

# 활동 순서 정렬 (정적인 것 -> 동적인 것 순으로 보면 예쁨)
order = ['lying', 'sitting', 'standing', 'walking', 'climbingup', 'climbingdown', 'running', 'jumping']

sns.boxplot(
    x='activity',
    y='svm',
    data=subset,
    order=order,
    palette="viridis",
    showfliers=False # 이상치(점) 제거하고 몸통만 깔끔하게 보기
)

plt.title(f'Activity Intensity Comparison (Position: {target_pos.upper()})', fontsize=20)
plt.ylabel('Acceleration Magnitude (m/s²)', fontsize=14)
plt.xlabel('Activity', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

plt.show()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# ---------------------------------------------------------
# 1. 세션별(사람+활동+Trial) 측정 시간 계산
# ---------------------------------------------------------
# (끝 시간 - 시작 시간) -> 분(Minute) 단위 변환
session_durations = master_df.groupby(['subject_id', 'activity', 'trial'])['time'].apply(
    lambda x: (x.max() - x.min()) / 1000 / 60
).reset_index(name='duration_min')

# ---------------------------------------------------------
# 2. Barplot 그리기 (평균)
# ---------------------------------------------------------
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# 시간 긴 순서대로 정렬
order = session_durations.groupby('activity')['duration_min'].mean().sort_values(ascending=False).index

# Barplot (검은 세로선은 신뢰구간/편차를 의미함)
ax = sns.barplot(
    x='activity',
    y='duration_min',
    data=session_durations,
    order=order,
    palette='coolwarm',
    edgecolor='black' # 테두리 진하게
)

# ---------------------------------------------------------
# 3. 가이드 라인 & 값 표시
# ---------------------------------------------------------
# 목표 시간 선 긋기
plt.axhline(y=10, color='red', linestyle='--', linewidth=2, label='Target: 10 min')
plt.axhline(y=1.7, color='blue', linestyle='--', linewidth=2, label='Target: ~1.7 min')

# 막대 위에 숫자 쓰기 (평균 시간)
for p in ax.patches:
    height = p.get_height()
    if height > 0: # 0보다 클 때만
        ax.text(
            p.get_x() + p.get_width() / 2.,
            height + 0.2,
            f'{height:.1f}m',
            ha="center", fontsize=12, fontweight='bold', color='black'
        )

plt.title('Average Recording Duration per Activity', fontsize=18, pad=15)
plt.xlabel('Activity', fontsize=14)
plt.ylabel('Duration (Minutes)', fontsize=14)
plt.legend(loc='upper right')
plt.ylim(0, 12) # Y축 12분까지 넉넉하게

plt.show()

# 수치로도 확인
print(session_durations.groupby('activity')['duration_min'].mean().round(2))
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch

# ---------------------------------------------------------
# [설정] 1번 사람, Thigh(허벅지) 센서, X축
# ---------------------------------------------------------
TARGET_ID = 1
POS = 'thigh'
AXIS = 'x'
FS = 50  # 샘플링 레이트 50Hz

# 비교할 활동 3대장
activities = ['walking', 'running', 'standing']
colors = ['green', 'red', 'blue']

plt.figure(figsize=(12, 6))

for i, act in enumerate(activities):
    # 데이터 가져오기 (메모리 아끼기 위해 일부만)
    subset = master_df[
        (master_df['subject_id'] == TARGET_ID) &
        (master_df['activity'] == act)
    ][f'acc_{POS}_{AXIS}'].values[:2000] # 약 40초 분량

    # [핵심] 주파수 변환 (Welch's method: 노이즈 줄여줌)
    # f: 주파수(Hz), Pxx: 파워(강도)
    f, Pxx = welch(subset, fs=FS, nperseg=256)

    # 그래프 그리기 (로그 스케일 아님, 선형 스케일로 피크 확인)
    plt.plot(f, Pxx, label=act.upper(), color=colors[i], linewidth=2)

# 꾸미기
plt.title(f'Frequency Analysis (FFT) - Position: {POS.upper()}', fontsize=16)
plt.xlabel('Frequency (Hz)', fontsize=14)
plt.ylabel('Power Spectral Density', fontsize=14)
plt.legend(fontsize=12)
plt.xlim(0, 10) # 0~10Hz 구간만 보면 됨 (사람 움직임은 보통 이 안임)
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
````

````python
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# ---------------------------------------------------------
# [설정] 데이터 샘플링 (활동별 500개씩만 뽑자)
# ---------------------------------------------------------
# 전체 330만개 다 쓰면 안됨! 활동별 500개면 충분히 경향성 보임
samples_per_act = 500

# 활동별로 랜덤 샘플링 (균형 맞추기)
subset_df = master_df.groupby('activity').apply(
    lambda x: x.sample(n=samples_per_act, random_state=42)
).reset_index(drop=True)

# ---------------------------------------------------------
# [전처리] 1. Feature 선택 & 2. 정규화 (Scaling)
# ---------------------------------------------------------
# acc_, gyro_ 컬럼만 선택
feature_cols = [c for c in subset_df.columns if 'acc_' in c or 'gyro_' in c]

X = subset_df[feature_cols].values
y = subset_df['activity'].values

# t-SNE는 거리 기반이라 스케일링 필수! (StandardScaler)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ---------------------------------------------------------
# [분석] t-SNE 돌리기 (약 1~2분 소요 예상)
# ---------------------------------------------------------
print("⏳ t-SNE 계산 중... (잠시만 기다려주세요)")
tsne = TSNE(n_components=2, perplexity=30, random_state=42, n_iter=1000)
X_embedded = tsne.fit_transform(X_scaled)

# ---------------------------------------------------------
# [시각화] 산점도 그리기
# ---------------------------------------------------------
# 시각화용 데이터프레임
tsne_df = pd.DataFrame(X_embedded, columns=['Component 1', 'Component 2'])
tsne_df['Activity'] = y

plt.figure(figsize=(12, 10))
sns.scatterplot(
    x='Component 1', y='Component 2',
    hue='Activity',
    palette='bright', # 색깔 뚜렷하게
    data=tsne_df,
    legend='full',
    alpha=0.7,
    s=60 # 점 크기
)

plt.title('t-SNE Visualization of Sensor Data (Raw Signals)', fontsize=20, pad=20)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````
